```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/diabetes.csv')
bmi_population = data['BMI']
percentile_population = bmi_population.quantile(0.98)
sample_size = 100
bmi_sample = bmi_population.sample(sample_size, random_state=42)
percentile_sample = bmi_sample.quantile(0.98)
plt.figure(figsize=(10, 7))
plt.hist(bmi_population, bins=30, alpha=0.7, label='Population', color='blue')
plt.axvline(percentile_population, color='brown', linestyle='dotted', linewidth=2, label='98th Percentile (Population)')
plt.hist(bmi_sample, bins=30, alpha=0.7, label='Sample', color='yellow')
plt.axvline(percentile_sample, color='green', linestyle='dashed', linewidth=2, label='98th Percentile (Sample)')
plt.title('BMI Distribution and 98th Percentile Comparison')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.legend()
plt.show()
print(f'98th Percentile of BMI (Population): {percentile_population}')
print(f'98th Percentile of BMI (Sample): {percentile_sample}')

```
