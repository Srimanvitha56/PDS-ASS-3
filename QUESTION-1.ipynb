```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
diabetes_data = pd.read_csv('/content/diabetes.csv')
np.random.seed(42)
sample_data = diabetes_data.sample(25)
average_glucose_sample = sample_data['Glucose'].mean()
highest_glucose_sample = sample_data['Glucose'].max()
average_glucose_population = diabetes_data['Glucose'].mean()
highest_glucose_population = diabetes_data['Glucose'].max()
labels = ['Avg Sample', 'Highest Sample', 'Avg Population', 'Highest Population']
values = [average_glucose_sample, highest_glucose_sample, average_glucose_population, highest_glucose_population]
plt.bar(labels, values, color=['red', 'yellow', 'blue', 'green'])
plt.ylabel('Glucose Data')
plt.title('Glucose Stats Sample & Population Vs Glucose Data')
plt.show()
```
